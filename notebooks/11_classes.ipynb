## Inheritance example

```python
class Polygon:
    def __init__(self, no_of_sides):   # it's a constructor
        self.n = no_of_sides
        self.sides = [0 for i in range(no_of_sides)]

    def set_sides(self, sides):
        self.sides = sides
        n = sides

    def show_sides(self):
        for i in range(self.n):
          print("Side", i+1, "is", self.sides[i])
```

```python
q = Polygon(3)
```

```python
q.show_sides()
```

```python
q.set_sides([1, 2, 3])
```

```python
q.show_sides()
```

```python
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self, 3)

    def get_area(self):
        a, b ,c = self.sides
        # calculate the semi-perimeter
        s = (a + b + c) / 2
        area = (s * (s-a)*(s-b)*(s-c)) ** 0.5 # jsut the formula
        return area
```

```python
t = Triangle()
t.n
```

```python
t.set_sides([3, 4, 5])
t.show_sides()
```

```python
t.get_area()
```

```python
class Rectangle(Polygon):
    def __init__(self):
        Polygon.__init__(self, 4)

    def set_sides(self, sides):
        assert sides[0] == sides[2], "Opposite sides must be equal"
        assert sides[1] == sides[3], "Opposite sides must be equal"
        self.sides = sides

    def get_area(self):
        area = self.sides[0] * self.sides[3]
        return area

```

```python
r = Rectangle()
```

```python
r.set_sides([1, 2, 3, 3])
```

```python
r.set_sides([1, 2, 1, 2])
```

```python
r.get_area()
```
