# Defining new function

```python
def square(x):
    return x * x
```

```python
square(3.4)
```

```python
square(56)
```

```python
def sum_squares(x,y):
    return square(x) + square(y)
```

```python
sum_squares(5,6)
```

```python
## default variabes
def sqrt(num, guess = 1):
    print ("Finding the square root of ", num, " starting with guess ", guess)
```

```python
sqrt(25)
```

```python
### defing funciton documentation
def pressure(v,t,n):
    """ Compute the pressure in pascals of an ideal gass.(This is called doc stirng).

    applies ideal gass Law.
    v -- volume of gass, in cubic meters
    t -- absolute temperature in degree kelvin
    n -- number of particals of gass
    """
    k = 1.38e-23
    return n * k * t / v
```

```python
### now we can see the above doc string of fucntion using help func
help(pressure)
```
