```python
## fucniton to count how many times a values repeats in a sequence
```

```python
def count(s,value):
    """Count how many times value occur in s."""
    count = 0
    for i in s:
        if i == value:
            count += 1

    return  count
        
```

```python
count(digits,8)
```
