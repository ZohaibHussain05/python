# Dictionaries
#### are simailar to Lists


```python
d = {
    "package":          "A package is a module that can be added to any program",
    "random" :          "Any data or information that has an order",
    "unirary operator": "An operator that takes only one value for its operation",
    "any":              2,  # so we can add any thing to the value like intiger, float string
    "someother":        ["meaning 1", "meaninig 2"]  # we can also add other structres like lists and tuples
}
```

```python
type(d)
```

```python
d["someother"]
```

### we can also assign a new variaible in a dict by assigment oeprator

```python
d["variable"] = "A memory location, the value of which can change"
```

### iterating over dict

```python
for a in d:
    v = d[a]
    print(a, "=", v)
```

```python
### finding wheter a key exists in the dict
'package' in d
```

```python
if 'else' not in d:   # we can move 'not' before the keyword
    print("The keyword 'else' is not in the d...")
else:
    "They keyword was found"
```

### dict.items()
##### return the list of tuples
#####  convert the dict into list of tupe. each tuple constainss a key value pair

```python
d.items() # return the list of tuples  
```

#### Now we can iterate over the dict same as we iterate
#### over the list of pairs in prev lec ie [ [1, 2] , [3 ,4] ]

```python
for k, v in d.items():
    print(k , "=", v)
```
