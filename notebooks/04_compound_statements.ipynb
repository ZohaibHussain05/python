```python
def is_weekday(day,month,year):
    import datetime
    return datetime.datetime(year=year,month=month,day=day).weekday() <= 4
```

```python
def good_enough(guess,x):
    if abs(guess * guess - x) < 0.1:
        return True
    else:
        return False 
```

```python
def improve_guess(guess,x):
    if guess < x:
        return guess + 0.1
    else:
        return guess - 0.1
```

```python
def sqrt(x, guess=0.0):
    # not dealing wiht complex numbers
    if x < 0:
        return None
    if good_enough (guess,x):
        return guess
    else:
        new_guess = improve_guess(guess,x)
        return sqrt(x,new_guess)
```

```python
sqrt(4,1.8)
```

```python
## lets improve the guess a little more
def average(a,b):
    return (a+b)/2.0
```

```python
def improve_guess(guess,x):
    if guess == 0:
        return 0.1
    better_guess = average(guess,x/guess)
    return better_guess
```
