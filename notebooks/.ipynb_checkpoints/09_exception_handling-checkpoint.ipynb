```python
def div(num, den):
    return num / float(den)
```

```python
# div by zero is expeceted so we can handle tha
```

```python
try:
    print("Div by 4 by 2 is:", div(4, 2))
    print("Div by 4 by 0 is:", div(4, 0))
except ZeroDivisionError:
    print("Do not divide by Zero.")

print("done")
```

### Bubbling exceptions

```python
def calculator(op, n1, n2):
    if( op == 'div'):
        return div(n1, n2)
    elif( op == 'mul'):
        return mul(n1, n2)
```

```python
print(calculator('div', 4, 0))
```

### now here the exception bubble up back to from function call stack
### if not handled in middle we can now handle it and do appropiate action based on our app type

```python
try:
    print(calculator('div', 4, 2))
    v
except ZeroDivisionError:
    print("Do not divide by 0.") # now user has better response
```

## Exception Hierarchy

#### now we have hierarchy of Exceptions and errors

```python
try:
    print(calculator('div', 4, 2))
    v
except ZeroDivisionError:
    print("Do not divide by 0.") # now user has better response
except NameError:
    print("That's a bad name")
```

#### we have all possibilites of exceptions in the hierarchy
#### we can handle all of them seperately but we can handle them all in above type in hierarchy 

```python
try:
    print(calculator('div', 4, 2))
    v
except Exception:
    print("No idea what went wrong") # hence, don't do this
```
