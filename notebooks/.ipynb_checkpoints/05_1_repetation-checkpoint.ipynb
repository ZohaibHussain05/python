```python
## sometimes we don't know how many times we have to iterate
## but we know when to stop. This is called while loop
## for example we have to print all multiples of 1.3 from 0 to 1
```

```python
n = 0.13

while n < 1:
    print(n)
    n += .13
```

```python
def factorial(n):
    fac = 1
    for i in range(2,n+1):
        fac*= i
    return fac
```

```python
factorial(5)
```

```python
def factorial(n):
    fac = 1
    i = 1
    while(i<=n):
        fac *= i
        i+=1
    return fac
```

```python
factorial(120)
```
