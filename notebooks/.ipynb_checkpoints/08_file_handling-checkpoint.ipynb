```python
directory = 'res'
```

```python
name = 'dummy.txt'
```

```python
import os
```

```python
filename = os.path.join(directory, name)
print(filename)
```

```python
os.getcwd()
```

```python
f = open(filename, 'r') # handle

for line in f:
    print(line)

f.close() # never forget this
```

#### there are blank lines that shouldn't be there
#### as each line of file also have line break and python print also a end line
#### now windows add \r\n for line break and *nix use \n. we leave it on python to handle and remove it

```python
f = open(filename, 'r') # handle

for line in f:
    print(line.strip()) ## strip func remove all white-spaces from both sides of string

f.close() # never forget this
```

### python also have context managers
#### so you have not to close files or resources 
#### python do house keeping itself

```python
with open(filename, 'r') as f:
    for line in f:
        print(line.strip())

##  so now no closing here.. now you can forget it
```

```python
with open(filename, 'r') as f:
    print(f.read())  # read the whole file
```

#### if we need each line in a separate list element we can do following

```python
'a-b-c'.split('-')
```

```python
l = []

with open(filename, 'r') as f:
    l = f.read().split('\n')
```
