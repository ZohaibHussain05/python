# Case Study: Comma Separated Values

```python
directory = 'res'
```

```python
import os

csv_filename = os.path.join(directory, 'datafile.csv')
print(csv_filename)
```

#### now we can use concept from prev lec to split the line in lists

```python
with open(csv_filename, 'r') as f:
    for line in f:
        print(line.strip().split(',')) # let's save this in a list
```

```python
rows = []

with open(csv_filename, 'r') as f:
    for line in f:
        rows.append(line.strip().split(','))
```

```python
rows
```

### we don't use first line its a header
### we have to convert all the string to floats to manipulate the data

```python
for r in rows[1:]:
    for i in range(2,6):
        r[i] = float(r[i])
```

```python
rows
```

# Writing csv files
### we have to join some string by a delimeter(,)

```python
'.'.join( ['a,', 'b', 'c'] ) # opposite of split
```

#### now after manipulation when we want to write data back to file 
#### we have to convert back to floats, we can use concept of list comprehendion

```python
for r in rows:
    print( [str(c) for c in r] )
```
