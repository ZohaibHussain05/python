```python
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n-1) + fib(n-2) ## but that not a good approach too many recursive calls
```

```python
%time fib(40)
```

```python
## now its very slow we have to improve it
## too many recursive calls we have to use dynamic programming 
```

```python
def fib2(n):
    if n <= 1: return n

    a , b = 0 , 1 

    for i in range(1,n):
        a , b = b , a + b 
        # python first place the evaluated value on right side then assign to left
        """
        n = a  + b
        a = b
        b = n
        """
    return n
```

```python
%time fib2(40)
```
