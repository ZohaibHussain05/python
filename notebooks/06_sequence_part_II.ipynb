# Nested Sequences

```python
## we can have anyother list at a index of a list
## hence we can form multi-dimenstional lists
```

```python
pairs = [
    [ 1, 20],
    [ 2, 30],
    [ 3, 40],
    [ 4, 50]
]
```

```python
pairs[1]
```

```python
for a in pairs:
    print(a)
```

```python
for x , y in pairs:
    print("x= ",x, "and y=" , y)
```

### Nested loops

```python
for j in range(1,5):
    print("Printing talbe of",j)
    for i in range(1,11):
        print(j, "x" , i , "=" , j * i)
```

### List Comprehensions

```python
nums = [1, 2, 3, 4, 5, 6, 7, 1000]
```

```python
# get lsit of all evens from nums list
evens = [i for i in nums  if i & 2 == 0]
print(evens)
```

```python
## get all factors of 12
[x for x in range(1,12+1) if 12 % x == 0]
```
