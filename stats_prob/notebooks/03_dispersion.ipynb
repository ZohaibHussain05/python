## Generating a LOT of numbers

```python
num_samples = 10000  # get rid of magic numbers
```

```python
uniform_vals = np.random.uniform(low=0.0, high=10.0, size=num_samples)
```

```python
sns.displot(uniform_vals, bins=20, kde=False)
plt.ylabel('Frequency')
plt.xlabel('Value')
plt.title("Uniform Values")
sns.despine(offset=10, trim=True) # move axes away
plt.show()
```

```python
normal_vals = np.random.normal(loc=5.0, scale=10.0, size=num_samples)
# More values closer to loc  .... sort of (1 5 5 5 5 7)
```

```python
sns.displot(normal_vals, bins=20, kde=False)
plt.ylabel('Frequency')
plt.xlabel('Value')
plt.title("Normal Values")
sns.despine(offset=10, trim=True) # move axes away
plt.show()
```

```python
print(f"Uniform values mean: {np.mean(uniform_vals)}")
print(f"Normal values mean:  {np.mean(normal_vals)}")
```

```python
print(f"Uniform values variance: {np.var(uniform_vals)}")
print(f"Normal values variance:  {np.var(normal_vals)}")
```

```python
# we can see mean for both types of data is same but varaince is change
print(f"Uniform values std: {np.std(uniform_vals)}")
print(f"Normal values std:  {np.std(normal_vals)}")
```

```python
normal_vals = np.random.normal(loc=5.0, scale=10.0, size=num_samples)

nv_mean = np.mean(normal_vals)
nv_sd = np.std(normal_vals)
```

```python
sns.displot(normal_vals, bins=20, kde=False)
plt.ylabel('Frequency')
plt.xlabel('Value')
plt.title("Uniform Values")
sns.despine(offset=10, trim=True) # move axes away

# plot the std deviateion line
x_c, y_c = ([nv_mean, nv_mean], [0, 1600])
plt.plot(x_c, y_c, color='green', linewidth=2)

x_c, y_c = ([nv_mean + nv_sd, nv_mean + nv_sd], [0, 1600])
plt.plot(x_c, y_c, color='red', linewidth=2)

x_c, y_c = ([nv_mean - nv_sd, nv_mean - nv_sd], [0, 1600])
plt.plot(x_c, y_c, color='red', linewidth=2)


plt.show()
```

#### in most of normal distrubuted data about 68% of data lies in one std deviation from mean.
#### and about 95% of data lies in 2 std deviation from the mean.

```python
# adding more dispersion to data
normal_vals = np.random.normal(loc=5.0, scale=15.0, size=num_samples)

nv_mean = np.mean(normal_vals)
nv_sd = np.std(normal_vals)
```

```python
sns.displot(normal_vals, bins=20, kde=False)
plt.ylabel('Frequency')
plt.xlabel('Value')
plt.title("Uniform Values")
sns.despine(offset=10, trim=True) # move axes away

# plot the std deviateion line
x_c, y_c = ([nv_mean, nv_mean], [0, 1600])
plt.plot(x_c, y_c, color='green', linewidth=2)

x_c, y_c = ([nv_mean + nv_sd, nv_mean + nv_sd], [0, 1600])
plt.plot(x_c, y_c, color='red', linewidth=2)

x_c, y_c = ([nv_mean - nv_sd, nv_mean - nv_sd], [0, 1600])
plt.plot(x_c, y_c, color='red', linewidth=2)


plt.show()
```

#### both graphs looks similar as both have same distance between the lines but there is catch!!! 
#### look at x axis values. Now lests **adjust the values at the x axis** using plt.xlim() then see the plots

```python
## now lests put them togather
normal_vals = np.random.normal(loc=5.0, scale=10.0, size=num_samples)
normal_vals2 = np.random.normal(loc=5.0, scale=15.0, size=num_samples)


def plot_dist(vals, label):

    nv_mean = np.mean(vals)
    nv_sd = np.std(vals)

    sns.displot(vals, bins=20, kde=False)

    plt.xlim(-60, 60)
    
    plt.ylabel('Frequency')
    plt.xlabel(label)
    plt.title("Uniform Values")
    sns.despine(offset=10, trim=True) # move axes away
    
    # plot the std deviateion line
    x_c, y_c = ([nv_mean, nv_mean], [0, 1600])
    plt.plot(x_c, y_c, color='green', linewidth=2)
    
    x_c, y_c = ([nv_mean + nv_sd, nv_mean + nv_sd], [0, 1600])
    plt.plot(x_c, y_c, color='red', linewidth=2)
    
    x_c, y_c = ([nv_mean - nv_sd, nv_mean - nv_sd], [0, 1600])
    plt.plot(x_c, y_c, color='red', linewidth=2)
    
    
    plt.show()

plot_dist(normal_vals, "Less Vareint")
plot_dist(normal_vals2, "More Varient")
```

### Now we can see the difference between the 2 pltos

```python
# plot the median income
med_inc = [x[0] for x in houses.data]
mi_mean = np.mean(med_inc)
mi_sd = np.std(med_inc)

sns.displot(med_inc, bins=20, kde=False)
plt.ylabel('Frequency')
plt.xlabel('Median Income')
plt.title("Uniform Values")
sns.despine(offset=10, trim=True) # move axes away

# plot the std deviateion line
x_c, y_c = ([mi_mean, mi_mean], [0, 4000])
plt.plot(x_c, y_c, color='green', linewidth=2)

x_c, y_c = ([mi_mean + mi_sd, mi_mean + mi_sd], [0, 4000])
plt.plot(x_c, y_c, color='red', linewidth=2)

x_c, y_c = ([mi_mean - mi_sd, mi_mean - mi_sd], [0, 4000])
plt.plot(x_c, y_c, color='red', linewidth=2)


plt.show()
```

#### we can see that the data is skewed

##### std deviation value does not tell us much
##### what is differnce between 10.7 and 1084 std deviation?
its depends upon the value of mean.
therefore we use another variable called cofficient of variation ie. "Relative measure of deviation".

COV = (std / mean ) * 100

10.7 std of 100 mean is 10.7% 
and
1094 std of 100,000 mean is only 1.09%
