## One Hot Encoding

```python
B = ['bird', 'cat', 'cat', 'dog', 'bird', 'bird']

d = {'categorical': B}

df = pd.DataFrame(d)
```

```python
df
```

```python
df['categorical']
```

```python
# "dummies" is used to create columns corresponding to unique values not comparable
dfDummies = pd.get_dummies(df['categorical'], prefix = 'category')
```

```python
dfDummies
```
