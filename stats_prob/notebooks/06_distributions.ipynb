## Distributions

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(color_codes=True)

import datetime
```

### Bernoulli Distribution

```python
from scipy.stats import bernoulli
```

not only scipy provide distributions. there are alot of other packages that also provide

```python
data_bern = bernoulli.rvs(size=10000, p=0.6)

ax = sns.histplot(data_bern,
                 bins=50,
                 kde=False,
                )
ax.set(xlabel='Bernoulli Distribution', ylabel='Frequency')
```

```python
print(data_bern)
```

### Binomial Distribution

```python
from scipy.stats import binom
```

```python
data_binom = binom.rvs(n=10, p=0.5, size=1000)
ax = sns.histplot(data_binom,
                 bins=100,
                 kde=False,
                )
ax.set(xlabel='Binomial Distribution', ylabel='Frequency')
```

```python
print(data_binom)
```

### Poisson Distribution

```python
from scipy.stats import poisson
```

```python
data_poisson = poisson.rvs(mu=3, size=10000)
ax = sns.histplot(data_poisson,
                 bins=50,
                 kde=False,
                )
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')
```

```python
print(data_poisson[:100])
```
