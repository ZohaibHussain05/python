## Data Visualization

```python
# lets assume some simple data
y = [6, 7, 2, 4, 5]
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline # display the plot in side the notebook cell instead of opening new window
```

```python
plt.plot(y)
plt.show()
```

```python
# create the plot but there is no labels
# eg plot title, axis label and unit

# create a figure 
plt.figure(figsize=(10,5))

# add a line renderer with legend and line thickness
plt.plot(y)

# now to make figure readable 
plt.title("First Example Plot") # title of plot
plt.xlabel("x vals (s)") # x axis label with seconds unit
plt.ylabel("y vals (m)") # y axis label with meter units
plt.grid() # draw grid in bg

plt.show()
```

```python
## now lets plot BMXWT also similarly

# create a figure 
plt.figure(figsize=(20,5))

# add a line renderer with legend and line thickness
x = range(da['BMXWT'].size)
y = da['BMXWT']

# make a scatter plot instad of line
plt.scatter(x, y) #, marker='x', s=0.1)

# now to make figure readable 
plt.title("Weights of Individulas") # title of plot
plt.xlabel("ID of Individulas") # x axis label with seconds unit
plt.ylabel("Weight (kg)") # y axis label with meter units

plt.show()
```

```python
## now lets plot BMXWT also similarly

# create a figure 
plt.figure(figsize=(20,5))

# add a line renderer with legend and line thickness
x = range(da['BMXWT'].size)
y = da['BMXWT']

# make a scatter plot instad of line
plt.scatter(x, y, marker='x', s=0.7)
# marker change dot to x and s change the size of dots

# now to make figure readable 
plt.title("Weights of Individulas") # title of plot
plt.xlabel("ID of Individulas") # x axis label with seconds unit
plt.ylabel("Weight (kg)") # y axis label with meter units

plt.show()
```
