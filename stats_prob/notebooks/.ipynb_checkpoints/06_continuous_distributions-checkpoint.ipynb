# Continuous Distribution

### Uniform Distribution

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(color_codes=True)

import datetime
```

```python
from scipy.stats import uniform
```

```python
n = 10000

n *= 100

start = 10
width = 20

data_uniform = uniform.rvs(size=n, loc=start, scale=width)

ax = sns.histplot(data_uniform,
                 bins=100,
                 kde=False,
                )
ax.set(xlabel='Unifrom Distribution', ylabel='Frequency')
```

### Noemal Distribution / Guassian Distribution

```python
from scipy.stats import norm
```

```python
data_normal = norm.rvs(size=10000, loc=0, scale=1)  # as mean = 0 and std dev. = 1, it's standard normal dist.

ax = sns.histplot(data_normal,
                 bins=100,
                 kde=True,
                )
plt.xlim(-10, 10)
ax.set(xlabel='Standard Normal Distribution', ylabel='Frequency')
```

```python
data_normal = norm.rvs(size=10000, loc=2, scale=2)

ax = sns.histplot(data_normal,
                 bins=100,
                 kde=True,
                )
plt.xlim(-10, 10)
ax.set(xlabel='Normal Distribution', ylabel='Frequency')
```

### Exponential Distribution

```python
from scipy.stats import expon
```

```python
data_expon = expon.rvs(scale=1, loc=0, size=1000)

ax = sns.histplot(data_expon,
                 bins=100,
                 kde=True,
                )
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```
