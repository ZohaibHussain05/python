# Multithreading

```python
import requests
```

```python
gists = ['https://gist.github.com/recluze/1d2989c7e345c8c3c542', 
        'https://gist.github.com/recluze/a98aa1804884ca3b3ad3', 
        'https://gist.github.com/recluze/5051735efe3fc189b90d', 
        'https://gist.github.com/recluze/460157afc6a7492555bb', 
        'https://gist.github.com/recluze/5051735efe3fc189b90d', 
        'https://gist.github.com/recluze/c9bc4130af995c36176d']
```

```python
from contextlib import contextmanager
import time

@contextmanager   # mark the function as 'Context Manager' so we can use it by 'with' keyword
def timeit():
    start_time = int(round(time.time() * 1000))

    yield         # this is from the generator\     
    
    end_time = int(round(time.time() * 1000))
    elapsed = end_time - start_time
    print(f"Code took {elapsed} ms to run.")
```

```python
def get_gist(g):
    print(f"Starting the request: {g}")

    r = requests.get(g)
    g_length = len(r.text)
    print(f"Got response of length: {g_length}")
    
```

```python
with timeit():
    
    for g in gists:
        get_gist(g)
    
    print("All Done!")
```

### Let's do all these in parallel!

```python
import threading

with timeit():
    threads = []
    for g in gists:
        # create worker thread for each gist
        t = threading.Thread(target=get_gist, args=(g, ))

        # Start the thread
        t.start()

        threads.append(t) # keep list of threads for record

    for t in threads:
        t.join()          # wait till t finishes

    print("All Done!")
```
