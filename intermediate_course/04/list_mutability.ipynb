# __List and Mutability__

```python
x = 34
y = x
x = 29
print(f"x = {x} , y = {y}")
```

```python
mammals = ['cat', 'kangroo',  'horse']
fish = ['tuna', 'shark', 'catfish']
```

```python
animals = [mammals, fish]
print(animals)
```

```python
mammals[0] = 'mule'
print(f"mammammals = {mammals}")
print(f"animals = {animals}")
```

can visualize this on python tuitor

```python
mammals_2 = mammals
mammals_2[0] = 'dog'
print(f"mammammals_2 = {mammals_2}")
print(f"mammammals = {mammals}")
print(f"animals = {animals}")
```

#### so till now it's just refrence of previous list. But what if we want hard copy of list, so we can change them seperately.

```python
mammals_3 = mammals[:] # this creates a copy of list!
```

```python
mammals_3[0] = 'mouse'
print(f"mammammals_3 = {mammals_3}")
print(f"mammammals = {mammals}")
print(f"animals = {animals}")
```

```python
s1 = ['student 1', 'James', 'Brown']
s2 = ['student 2', 'Chris', 'Black']
s3 = ['student 3', 'Phil', 'Blue'] 
```

```python
students = [s1, s2, s3]
print(students)
```

```python
students_copy = students[:]
print(students_copy)
```

```python
students_copy[0][0] = 'student 1 changed'
print(students_copy)
print(students)
```

so we have spread the list of list into the students copy it make a shallow copy which ultimately putt the inner list objects (refrence) into\
new "students_copy" variable. which is not a deep copy.

## __Deep Copying__

```python
import copy
```

```python
students_3 = copy.deepcopy(students)
```

```python
students_3[0][0] = 'student 1 changed again'
print(students_3)
print(students)
```
