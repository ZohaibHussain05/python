# Decorators

Recall the typical fib method. 

$$0, 1, 1, 2, 3, 5, 8, \ldots $$

$$\textrm{fib}(0) = 0$$
$$\textrm{fib}(1) = 1$$

$$\textrm{fib}(n) = \text{fib}(n-2) + \text{fib}(n-1)$$

```python
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n-2) + fib(n-1)
```

```python
%time fib(40)
```

This will take a bit of time so let's see what's wrong. 
when we make function call tree we can see that fabonacci of same number is being called many times.
This lead to the problem of repeated sum, which is usually solved by the complex dynamic programming apporach.

But we can use concept of higher-order fucntions to tackle this issue. But let's step back for a minute.

```python
def logger(f):

    # f will be "remembered" by the wrapper function even after we exit logger.
    # This is concept of closure

    def wrapper(n):
        print("I'm going to call a funciton")
        v = f(n)
        print("The function returned value:", v)
        return v

    return wrapper
```

```python
logged_fib = logger(fib)
```

```python
logged_fib(4)
```

Now that we can do stuff beforethe fib call, let's see if we can save some values that repearedly needed.

```python
def memoize(f):
    mem = {}

    def memoized_function(n):
        if n not in mem:
            mem[n] =f(n)

        return mem[n]
        
    return memoized_function
```

```python
fib = memoize(fib)
```

```python
%time fib(40)
```

# Syntactic Sugar

we can write it in a short way

```python
def memoize(f): # same previous function
    mem = {}

    def wrapper(n):
        if n not in mem:
            mem[n] =f(n)

        return mem[n]
        
    return wrapper
```

```python
@memoize                       # this is called a "decorator", EQUALS: fib = memoize(fib)
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n-2) + fib(n-1)
```

```python
fib(50)
```
