# Simple Example

Let's say we want to write a function for calculating sums: $$\sum_{i=1}^{10}{i^2}$$

```python
def summation(low, high):
    total = 0
    for i in range(low, high + 1):
        val = i ** 2
        total += val

    return total
```

```python
summation(1, 3)
```

Now, if we have to write another function that does this:$$\sum_{i=1}^{10}{i^3}$$

```python
def summation(low, high):
    total = 0
    for i in range(low, high + 1):
        val = i ** 3
        total += val

    return total
```

But what's the difference? Only the `val` has changed. Can't we just write one function and have you decide how val needs to be calculated? 

```python
def square(x): 
    return x ** 2 
def cube(x): 
    return x ** 3 
```

```python
def summation(low, high, fn):
    total = 0
    for i in range(low, high + 1):
        val = fn(i)
        total += val

    return total
```

Now we can call summation and just change the function that calculated val.

```python
summation(1, 2, square)
```

```python
summation(1, 2, cube)
```

we don't even have to name the funcitons!

```python
summation(1, 10, lambda i: i**2 ) # this is an anonymous funciton.
```

Notice that this statement now looks very similar to the actual math notation we had earlier:  

$$ 
\sum_{i=1}^{10}{i^2}
$$

So, if you have to write another one for this:  $$\sum_{i=1}^{10}{2i^2}$$ 

```python
summation(1, 10, lambda i: 2 * (i ** 2))
```

Summation is a high order function because it takes antoher function as input

# Case Study: square Roots

```python
def sqrt(x, guess=0.1): 
    print("Trying:", guess, "-- Value:", guess*guess)
    if good_enough(guess, x): 
        return guess 
    
    else: 
        guess = improve_guess(guess, x)  
        return sqrt(x, guess)
```

```python
# Defined by the weather people 
def good_enough(guess, x): 
    if abs(guess * guess - x) < 1: 
        return True
    else: 
        return False 
```

```python
def avg(a, b): 
    return (a + b) / 2.0

def improve_guess(guess, x): 
    return avg(guess, float(x)/guess)
```

```python
sqrt(36)
```

Instead of hardcoding concept of good enough in `good_enough()` funciton, we make it dynamic

```python
def sqrt(x, is_ge=good_enough, guess=0.1): 
    print("Trying:", guess, "-- Value:", guess*guess)
    if is_ge(guess, x): 
        return guess 
    
    else: 
        guess = improve_guess(guess, x)  
        return sqrt(x, guess)
```

```python
# By the nuclear reactor people
def very_accurate_good_enough(guess, x):
    return abs(guess * guess - x) < 0.000000001
```

```python
sqrt(36, is_ge=very_accurate_good_enough)
```
