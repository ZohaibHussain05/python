#  Regular Expressions

```python
text = """03-02-2025 10:22pm: My email address is testuser@gmail.com and I check
          it often. You should send me an email!"""
```

We want to extract the email address from this string. 

But let's start small. 

```python
haystack = "Date: 03-02-2025"
```

```python
needle = "Da"
```

```python
import re
```

```python
re.match(needle, haystack)
```

```python
match = re.match(needle, haystack)
```

```python
match.group(0)
```

```python
needle = "Date: \d"
```

```python
res = re.match(needle, haystack)

if res:
    print("Found!")
    print(res.group(0))
else:
    print("Not Found!")
```

```python
needle = "......\d\d"
res = re.match(needle, haystack)

if res:
    print("Found!")
    print(res.group(0))
else:
    print("Not Found!")
```

```python
needle = ".{5}\s\d{2}"
res = re.match(needle, haystack)

if res:
    print("Found!")
    print(res.group(0))
else:
    print("Not Found!")
```

```python
needle = ".*\d{2}" # '*' means not sure how many instances
res = re.match(needle, haystack)

if res:
    print("Found!")
    print(res.group(0))
else:
    print("Not Found!")
```

But that's not what we expected! 

The problem is that `.*` is matching everything and `d{2}` matches the last two digits!

```python
needle = ".*?\d{2}" # '*' means not sure how many instances
res = re.match(needle, haystack)

if res:
    print("Found!")
    print(res.group(0))
else:
    print("Not Found!")
```

The `?` means that it should not be <strong>greedy</strong>. As soon as `\d{2}` can be satisfied, the effect of `.*` should be stopped. 

# Problem: This is already getting too complex!

Oh, but regular expressions look <strong>ugly</strong>!

That's because you need to approach them in a modular way, just as we break down our whole program into functions.  

```python
str_date = ".*?"   # can also use: ".*\s"
str_day = "\d{2}"

needle = str_date + str_day
res = re.match(needle, haystack)

if res:
    print("Found!")
    print(res.group(0))
else:
    print("Not Found!")
```

Let's take the more complicated string.

```python
haystack = text
print(haystack)
```

```python
str_day = "\d{2}-\d{2}-\d{4}"
str_time = "\d{2}:\d{2}pm"

needle = str_day + '\s' + str_time
res = re.match(needle, haystack)

if res:
    print("Found!")
    print(res.group(0))
else:
    print("Not Found!")
```

But what if time is in morning

```python
haystack = """03-02-2025 10:22am: My email address is testuser@gmail.com and I check
          it often. You should send me an email!"""

str_day = "\d{2}-\d{2}-\d{4}"
str_time = "\d{2}:\d{2}[ap]m"

needle = str_day + '\s' + str_time
res = re.match(needle, haystack)

if res:
    print("Found!")
    print(res.group(0))
else:
    print("Not Found!")
```

```python
haystack = """03-02-2025 10:22PM: My email address is testuser@gmail.com and I check
          it often. You should send me an email!"""

str_day = "\d{2}-\d{2}-\d{4}"
str_time = "\d{2}:\d{2}[apAP][mM]" # can handle any combination of am , pm

needle = str_day + '\s' + str_time
res = re.match(needle, haystack)

if res:
    print("Found!")
    print(res.group(0))
else:
    print("Not Found!")
```

```python
haystack = """03-02-2025 10:22pm: My email address is testuser@gmail.com and I check
          it often. You should send me an email!"""
```

```python
str_prefix = ".*"  # any thing

str_username = "[a-zA-Z0-9.]*"
str_domain = ".*\..*?"

str_email = str_username + '@' + str_domain

needle = str_prefix + "\s" + str_email + "\s"
res = re.match(needle, haystack)

if res:
    print("Found!")
    print(res.group(0))
else:
    print("Not Found!")
```

# Extraction of Email Adress

```python
# Add () around  stuff you need to extract
str_prefix = ".*"  # any thing

str_username = "([a-zA-Z0-9.]*)"
str_domain = "(.*\..*?)"

str_email = "(" + str_username + '@' + str_domain + ")"

needle = str_prefix + "\s" + str_email + "\s"
res = re.match(needle, haystack)

if res:
    print("Found!")
    print(res.group(1))
else:
    print("Not Found!")
```

```python
print(res.group(0))
print(res.group(1))
print(res.group(2))
print(res.group(3))
```
