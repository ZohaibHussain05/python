## Task2

```python
while True:
    subjects = 7
    while subjects < 1 or subjects > 6:
        print("Enter No of Subjects:")
        subjects = int(input())
    
    marks = []
    for i in range(subjects):
        print(f"Enter marks for {i+1} subject:")
        
        subjMarks = 101
        while subjMarks > 100 or subjMarks < 0: # valid marks check
            print("Please Enter valid marks[0-100]:")
            subjMarks = int(input())
            
        marks.append(subjMarks)
    
    totalMarks = sum(marks)
    averageMarks = totalMarks / subjects
    
    highest = max(marks)
    lowest = min(marks)
    
    grade = '-'
    if averageMarks >= 85:
        grade = 'A'
    elif averageMarks >= 70:
        grade = 'B'
    elif averageMarks >= 55:
        grade = 'C'
    elif averageMarks >= 40:
        grade = 'D'
    else:
        grade = 'F'
    
    status = "FAIL"
    if averageMarks >= 40 and lowest >= 33:
        status = "PASS"
    
    print("============================================")
    print("\t Grade Calculator")
    print("============================================")
    print(f"Marks Entered: {marks}")
    
    
    print("--------------------------------------------")
    print("\tResult Summary")
    print("--------------------------------------------")
    print(f"Total Marks: \t{totalMarks}")
    print(f"Average Marks: \t{averageMarks}")
    print(f"Highest Marks: \t{highest}")
    print(f"Lowest Marks: \t{lowest}")
    print("\n")
    print(f"Grade: \t\t{grade}")
    print(f"Status: \t{status}")
    print("============================================")

    print("Check another student? (yes/no):")
    if input().lower() == "no":
        break
```

## Task 3

### Part A

```python
noOfRooms = 9
while noOfRooms < 3 or noOfRooms > 8:
    noOfRooms = int(input("Enter No of Rooms:"))

roomsState = [False for i in range(noOfRooms)]
for i, r in enumerate(roomsState):
    state = input(f"Enter Room {i + 1} state:")
    if state.lower() == "clean":
        roomsState[i] = True
    else:
        roomsState[i] = False
## displaying all initial state
for state in roomsState:
    status = "Clean" if state else "Dirty" # as true in state represent the clean
    print(status, end = ', ')
```

### Part B

```python
def check_room(status):
    state = "MOVE" if status else "CLEAN" # move if room is clean else clean the dirty room
    return state
def count_dirty(rooms_list):
    count = 0
    for state in rooms_list:
        if state != True:
            count += 1

    return count
```

#### Part C

```python
def clean_rooms(rooms_list):
    for i, state in enumerate(rooms_list):
        print(f"Robot Entering room No: {i + 1}")
        currStatus = "clean" if state else "dirty"
        print(f"Status: {currStatus}")
        ## taking action
        action = check_room(state)
        print(f"Action: {action}")
        if state:
            print(f"Room {i + 1} stays clean\n")
        else:
            print(f"Room {i + 1} is now clean\n")

clean_rooms(roomsState)        
    
    
    
```

### Part D

```python
summary = {
    "Total Rooms"  : len(roomsState),
    "Rooms Cleaned": count_dirty(roomsState),
    "Already Cleaned":  len(roomsState) - count_dirty(roomsState)
}

print(summary)
```

```python
## printing summary

print("============================================")
print("\t ROOM CLEANING ROBOT")
print("============================================")

print("INITIAL STATE:")
print("--------------------------------------------")

for i in range(0, len(roomsState)):
        print(f"Room {i + 1}: {"clean" if roomsState[i] else "dirty"}")

print("\nDirty rooms to clean:", count_dirty(roomsState), '\n')

print("============================================")
print("ROBOT STARTING...")
print("============================================")
clean_rooms(roomsState)

print("============================================")
print("SUMMARY REPORT")
print("============================================")

for key, value in summary.items():
    print(key, ":", value)
print("============================================")
```
