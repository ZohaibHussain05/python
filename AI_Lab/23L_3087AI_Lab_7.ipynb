### Part B — Pandas

Load the full table into a Pandas DataFrame, then:
1. Filter and display only participants who solved **5 or more problems**
2. Sort the full DataFrame by `Problems` descending, then by `Penalty` ascending — this gives the final contest ranking
3. After sorting, reset the index and add a new column `Rank` numbered 1 through 7 so the ranking matches the sorted order
4. Using **`.iloc[]`**, retrieve the row at index position `2` from the ranked DataFrame, which corresponds to the **3rd place** finisher


```python
import pandas as pd

data = {
    'Participant': ['Ali Raza', 'Sara Khan', 'Ahmed Butt', 'Fatima Malik',
                    'Hassan Ali', 'Zara Siddiqui', 'Bilal Ahmed'],
    'Department':  ['CS', 'CS', 'SE', 'CS', 'AI', 'DS', 'CS'],
    'Problems':    [5, 7, 4, 7, 6, 3, 5],
    'Penalty':     [120, 85, 145, 90, 110, 160, 130],
    'Year':        ['3rd', '4th', '2nd', '3rd', '3rd', '1st', '2nd']
}

df = pd.DataFrame(data)

# 1. Filter: Problems >= 5
filtered = df[df['Problems'] >= 5]
print("Participants with 5+ problems solved:")
print(filtered)

# 2. Sort by Problems (desc) then Penalty (asc)
ranked_df = df.sort_values(by=['Problems', 'Penalty'], ascending=[False, True])
print("\nSorted Rankings:")
print(ranked_df)

# 3. Add Rank column (1 through 7) — reset index first
df = pd.DataFrame(ranked_df)
df["Rank"] = range(1, len(df) + 1)
# 4. Retrieve 3rd place using .iloc[]
third_place = df.iloc[2]
print("\n3rd Place Finisher:")
print(third_place)
```
