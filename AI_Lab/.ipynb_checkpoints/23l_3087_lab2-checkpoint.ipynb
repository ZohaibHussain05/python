## Q1. Ticket Pricing (5 Marks)

Given an int `age` of a visitor, return a string of the form `'Ticket price: <price>'`.

- Age **12 or under** → price is **5**
- Age **13 to 59** → price is **10**
- Age **60 or above** → price is **7**

Example: `ticket(8)` → `'Ticket price: 5'`, `ticket(65)` → `'Ticket price: 7'`

````python
def ticket(age):
    price = 0
    if(age < 0):
        return "Invalid age!!"
    if age <= 12:
        price = 5
    elif age <=59:
        price = 10
    else:
        price = 7
    return f"Ticket price: {price}."
        
````

````python
# Test
print(ticket(8))    # Expected: Ticket price: 5
print(ticket(30))   # Expected: Ticket price: 10
print(ticket(65))   # Expected: Ticket price: 7
````

## Q2. Middle Characters (5 Marks)

Given a string `s`, return a new string made of the middle three characters of the original string.

- If string length is **odd**, take the middle character and one on each side.
- If string length is **even**, take the two middle characters and one to the left.
- If string length is **less than 3**, return the string unchanged.

Example: `'Artificial'` → `'fic'`, `'abcdefg'` → `'cde'`

````python
def middle_chars(s):
    size = len(s)
    if size < 3:
        return s
    res = ''
    if size % 2: # length is odd
        return s[size // 2 - 1: size // 2 + 2]
    else:
        return s[size // 2 - 1: size // 2 + 2]
````

````python
# Test
print(middle_chars('Artificial'))  # Expected: fic
print(middle_chars('abcdefg'))     # Expected: cde
print(middle_chars('ab'))          # Expected: ab
````

## Q3. Vowel Replacer (5 Marks)

For a given string `s`, create a new string where every vowel (a, e, i, o, u) is replaced with `'#'`. The replacement should be case-insensitive (both `'A'` and `'a'` are replaced).

Example: `'Intelligence'` → `'#nt#ll#g#nc#'`

**Hint:** You can iterate through each character and build a new string.

````python
def vowel_replacer(s):
    new_s = ''
    vowels = ['a', 'e', 'i', 'o', 'u']
    for c in s:
        if c.lower() in vowels:
            new_s += '#'
        else:
            new_s += c
    return new_s
````

````python
# Test
print(vowel_replacer('Intelligence'))  # Expected: #nt#ll#g#nc#
````

## Q4. String Weaver (5 Marks)

Given two input strings `a` and `b`, create a new string by interleaving their characters (first char of `a`, first char of `b`, second char of `a`, second char of `b`, ...). If one string is longer, append the remaining characters at the end.

Example: `'abc', 'xyz'` → `'axbycz'`, `'hi', 'world'` → `'hwiorld'`

````python
def string_weaver(a, b):
    min_size = min(len(a), len(b))
    res = ''
    i = 0
    for i in range(0, min_size):
        res += a[i] + b[i]
    i += 1       # move i to min_size + 1
    if i < len(a):
        res += a[i:]
    if i < len(b):
        res += b[i:]
    return res
````

````python
# Test
print(string_weaver('abc', 'xyz'))    # Expected: axbycz
print(string_weaver('hi', 'world'))   # Expected: hwiorld
````

## Q5. Count Uppercase Words (5 Marks)

Given a list of strings, return the count of strings that start with an uppercase letter. Only consider strings with length 1 or more.

Example: `['Hello', 'world', 'AI', 'lab']` → `2`

````python
def count_uppercase(lst):
    count = 0
    for word in lst:
        if word[0].isupper():
            count += 1
    return count
````

````python
# Test
print(count_uppercase(['Hello', 'world', 'AI', 'lab']))  # Expected: 2
````

## Q6. Group and Sort by Length (5 Marks)

Given a list of strings, return a new list sorted by string length (shortest first). If two strings have the same length, they should appear in alphabetical order.

Example: `['banana', 'pie', 'kiwi', 'fig', 'apple']` → `['fig', 'pie', 'kiwi', 'apple', 'banana']`

**Hint:** Use a custom key function with `sorted()`.

````python
def compare(str1, str2): # comparing two strings
    if len(str1) != len(str2):
        return len(str1) < len(str2)
    for i in range(0, len(str1)):
        if(str1[i] == str2[i]):
            continue
        else:
            return str1[i] < str2[i]
    return True

def sort_by_length(lst):
    return sorted(lst, key=lambda x: (len(x), x))
````

````python
# Test
print(sort_by_length(['banana', 'pie', 'kiwi', 'fig', 'apple']))
# Expected: ['fig', 'pie', 'kiwi', 'apple', 'banana']
````

## Q8. Palindrome Detection (5 Marks)

Given a string `s`, create a function to detect if it is a palindrome. A palindrome reads the same backward as forward. Ignore spaces and case when checking.

Example: `'Race Car'` → `True`, `'hello'` → `False`

````python
def is_palindrome(s):
    s = s.lower()
    s = s.replace(' ','')
    return s == s[::-1]
````

````python
# Test
print(is_palindrome('Race Car'))  # Expected: True
print(is_palindrome('hello'))     # Expected: False
````

## Q9. List Flattening (5 Marks)

Given a nested list of integers, create a function to flatten the list to a single-level list. Your function should handle any level of nesting.

Example: `[[1, [2, 3]], [4, [5, [6]]]]` → `[1, 2, 3, 4, 5, 6]`

````python
def flatten(myList):
    if not isinstance(myList, list): # if not a list return single element list
        return [myList]
    res = []
    for lst in myList:
        res.extend(flatten(lst)) ## flat all list and add them to a single
    return res
````

````python
# Test
print(flatten([[1, [2, 3]], [4, [5, [6]]]]))  # Expected: [1, 2, 3, 4, 5, 6]
````

## Q10. List Intersection (5 Marks)

Given two lists of integers, return a sorted list containing only the elements common to both lists (without duplicates). Your program should work with lists of different sizes.

Example: `[1, 2, 2, 3, 4], [2, 3, 3, 5]` → `[2, 3]`

````python
def list_intersection(lst1, lst2):
    lst1 = sorted(lst1)
    lst2 = sorted(lst2)

    res = []
    i, j = 0, 0
    while i < len(lst1) and j < len(lst2):
        if(lst1[i] == lst2[j]):
            res.append(lst1[i])
            i += 1
            j += 1
        elif(lst1[i] < lst2[j]):
            i += 1
        else:
            j += 1

    return res
````

````python
# Test
print(list_intersection([1, 2, 2, 3, 4], [2, 3, 3, 5]))  # Expected: [2, 3]
````

## Q11. Character Frequency Counter (5 Marks)

Given a string `s`, return a dictionary containing the frequency count of each character in the string. Ignore spaces and treat uppercase and lowercase as the same character.

Example: `'Hello World'` → `{'h':1, 'e':1, 'l':3, 'o':2, 'w':1, 'r':1, 'd':1}`

````python
def char_frequency(s):
    s = s.lower()
    s = s.replace(' ','')
    res = {}
    for c in s:
        res[c] = res.get(c, 0) + 1
    return res
````

````python
# Test
print(char_frequency('Hello World'))
# Expected: {'h': 1, 'e': 1, 'l': 3, 'o': 2, 'w': 1, 'r': 1, 'd': 1}
````

## Q12. Unique Values in a Dictionary (5 Marks)

Given a dictionary, return a list of all the values in the dictionary without any duplicates. The returned list should be sorted.

Example: `{'a': 3, 'b': 1, 'c': 3, 'd': 2}` → `[1, 2, 3]`

````python
def unique_values(d):
    res = set()
    for value in d.values():
        res.add(value)

    return list(res)
````

````python
# Test
print(unique_values({'a': 3, 'b': 1, 'c': 3, 'd': 2}))  # Expected: [1, 2, 3]
````

## Q13. Most Frequent Value Keys (5 Marks)

Given a dictionary, return the key(s) of the value(s) that appear most often. If there is a tie, return all such keys as a sorted list.

Example: `{'a':1, 'b':2, 'c':1, 'd':3, 'e':2}` → `['a','b','c','e']` (1 and 2 both appear twice)

````python
def most_frequent_keys(d):
    value_fq = {}

    for value in d.values():
        if value in value_fq.keys():
            value_fq[value] += 1
        else:
            value_fq[value] = 0

    # sort the freq map
    value_fq = dict(sorted(value_fq.items(), key=lambda item: item[1], reverse=True))

    max_values = []
    max = next(iter(value_fq.values())) # frist element of key_fq
    
    for key,value in value_fq.items():
        if(value < max):
            break
        max_values.append(key)

    res = []
    for key, value in d.items():
        if value in max_values:
            res.append(key)
    return res
            
````

````python
# Test
print(most_frequent_keys({'a':1, 'b':2, 'c':1, 'd':3, 'e':2}))
# Expected: ['a', 'b', 'c', 'e']
````

## Q14. Combine Dictionaries (5 Marks)

Given two dictionaries `d1` and `d2`, return a new dictionary containing all keys and values from both. If a key exists in both, sum the values instead of overwriting.

Example: `{'a':1, 'b':2}, {'b':3, 'c':4}` → `{'a':1, 'b':5, 'c':4}`

````python
def combine_dicts(d1, d2):
    res = {}
    for key, value in d1.items():
        res[key] = res.get(key, 0) + value

    for key, value in d2.items():
        res[key] = res.get(key, 0) + value
    return res
````

````python
# Test
print(combine_dicts({'a':1, 'b':2}, {'b':3, 'c':4}))
# Expected: {'a': 1, 'b': 5, 'c': 4}
````

## Q15. Filter List of Dictionaries (5 Marks)

Given a list of dictionaries and a key-value pair, return a new list containing only the dictionaries where the given key has the given value.

Example:
```
[{'name':'Ali','age':20}, {'name':'Sara','age':22}, {'name':'Ali','age':25}]
key='name', value='Ali'
→ [{'name':'Ali','age':20}, {'name':'Ali','age':25}]
```

````python
def filter_dicts(lst, key, value):
    res_lst = []
    for dic in lst:
        for k, v in dic.items():
            if k == key and v == value:
                res_lst.append(dic)
                break

    return res_lst
````

````python
# Test
data = [{'name':'Ali','age':20}, {'name':'Sara','age':22}, {'name':'Ali','age':25}]
print(filter_dicts(data, 'name', 'Ali'))
# Expected: [{'name': 'Ali', 'age': 20}, {'name': 'Ali', 'age': 25}]
````

Given two sets of integers, create a function to return a tuple of two sets: the union and the intersection. The union contains elements in either set, while the intersection contains elements in both sets.

## Q18. Find Duplicates Using Sets (5 Marks)

Given a list of integers, return a sorted list of elements that appear more than once. Use sets to solve this problem efficiently.

Example: `[4, 3, 2, 7, 8, 2, 3, 1]` → `[2, 3]`

````python
def find_duplicates(lst):
    seen = set()
    dup = set()
    for x in lst:
        if x not in seen:
            seen.add(x)
        else:
            dup.add(x)
    return list(dup)
````

````python
# Test
print(find_duplicates([4, 3, 2, 7, 8, 2, 3, 1]))  # Expected: [2, 3]
````

## Q19. String Anagram Detection (5 Marks)

Given two strings, create a function to detect if they are anagrams of each other. An anagram is formed by rearranging all letters of another word. Ignore spaces and case.

Example: `'Cinema', 'Iceman'` → `True`

````python
def is_anagram(s1, s2):
    fc_s1, fc_s2 = {}, {}
    
    for c in s1:
        c = c.lower()
        fc_s1[c] = fc_s1.get(c, 0) + 1

    for c in s2:
        c = c.lower()
        fc_s2[c] = fc_s2.get(c, 0) + 1
        
    for key, value in fc_s1.items():
        if fc_s2.get(key, 0) != value:
            return False

    return True
````

````python
# Test
print(is_anagram('Cinema', 'Iceman'))  # Expected: True
print(is_anagram('Hello', 'World'))    # Expected: False
````

## Q20. Binary Search (5 Marks)

Given a sorted list of integers and a target value, create a function to perform binary search and return the index of the target element. If the target is not found, return `-1`. You must implement the algorithm yourself (**do not use built-in search methods**).

````python
def binary_search(lst, target, s, e):
    if s > e:
        return -1
    m = s + (e - s) // 2
    if lst[m] < target:
        return binary_search(lst, target, m + 1, e)
    elif lst[m] > target:
        return binary_search(lst, target, s, m - 1)
        
    return m
    
````

````python
# Test
print(binary_search([1, 3, 5, 7, 9, 11], 7, 0, 6))   # Expected: 3
print(binary_search([1, 3, 5, 7, 9, 11], 4, 0, 6))   # Expected: -1
````

## Q21. Merge Sort (5 Marks)

Given a list of integers, create a function to sort the list using the **merge sort** algorithm. Merge sort recursively breaks down the list into smaller sub-lists, sorts them, and then recombines them in sorted order.

````python
def merge_lsts(lst, s, m, e):
    A = [lst[i] for i in range(s, m)]
    B = [lst[i] for i in range(m, e)]

    i, j, k = 0, 0, s
    while i < len(A) and j < len(B):
        if(A[i] <= B[j]):
            lst[k] = A[i]
            i += 1
            k += 1
        else:
            lst[k] = B[j]
            j += 1
            k += 1

    while i < len(A):
        lst[k] = A[i]
        i += 1
        k += 1
    while j < len(B):
        lst[k] = B[j]
        j += 1
        k += 1

    return
        

def merge_sort(lst, s , e):
    if e - s <= 1:
        return 
    m = s + (e - s) // 2
    
    merge_sort(lst, s, m )
    merge_sort(lst, m, e)

    merge_lsts(lst, s, m, e)

    return lst
    
````

````python
# Test
print(merge_sort([38, 27, 43, 3, 9, 82, 10], 0, 7))  # Expected: [3, 9, 10, 27, 38, 43, 82]
````

## Q22. Quick Sort (5 Marks)

Given a list of integers, create a function to sort the list using the **quick sort** algorithm. Quick sort selects a pivot element and partitions the remaining elements into two sub-lists: those less than the pivot and those greater than the pivot, then recursively sorts each sub-list.

````python
def partition(lst, s, e):
    pivot = lst[e - 1]      
    i = s
    for j in range(s, e - 1):
        if lst[j] <= pivot:
            lst[i], lst[j] = lst[j], lst[i]
            i += 1
    # placing the pivot
    lst[i], lst[e - 1] = lst[e - 1], lst[i]

    return i


def quick_sort(lst, s, e):
    if e - s <= 1:
        return

    p = partition(lst, s, e)

    quick_sort(lst, s, p)
    quick_sort(lst, p + 1, e)

    return lst
````

````python
# Test
print(quick_sort([38, 27, 43, 3, 9, 82, 10], 0, 7))  # Expected: [3, 9, 10, 27, 38, 43, 82]
````
