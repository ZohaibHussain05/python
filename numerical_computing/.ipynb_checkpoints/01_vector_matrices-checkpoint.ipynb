# 1.1. Vector Representation

```python
class Vector:
    def __init__(self, x = 0.0, y = 0.0):
        self.x = x
        self.y = y

    def __str__(self):
        return f"[{self.x}, {self.y}]"
```

```python
a = Vector(2, 4)
```

```python
print(a)
```

```python
b = Vector(5, 2)
print(b)
```

```python
def add(self, b):
    c = Vector()
    c.x = self.x + b.x
    c.y = self.y + b.y
    return c

Vector.add = add
```

```python
c = a.add(b)
print(c)
```
