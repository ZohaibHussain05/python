# 2. Misc Operations with Numpy

```python
import math
def basic_sigmoid(x):
    """
    Compute sigmoid of x.

    Arguments:
    x -- A scalar

    Returns:
    s -- sigmoid(x)
    """

    s = 1. / ( 1. + math.e ** (-x))

    return s
```

```python
basic_sigmoid(-1)
```

```python
basic_sigmoid(0)
```

```python
basic_sigmoid(math.inf)
```

```python
x = [-1, 0, 3]
basic_sigmoid(x) # Why
```

```python
x = np.array(x)
basic_sigmoid(x)
```

```python
x = [-1, 0, 3]
np.apply_along_axis(basic_sigmoid, arr=x, axis=0)
```
