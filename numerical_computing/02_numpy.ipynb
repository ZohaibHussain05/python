# 2. Misc Operations with Numpy

```python
import math
def basic_sigmoid(x):
    """
    Compute sigmoid of x.

    Arguments:
    x -- A scalar

    Returns:
    s -- sigmoid(x)
    """

    s = 1. / ( 1. + math.e ** (-x))

    return s
```

```python
basic_sigmoid(-1)
```

```python
basic_sigmoid(0)
```

```python
basic_sigmoid(math.inf)
```

```python
x = [-1, 0, 3]
basic_sigmoid(x) # Why
```

```python
x = np.array(x)
basic_sigmoid(x)
```

```python
x = [-1, 0, 3]
np.apply_along_axis(basic_sigmoid, arr=x, axis=0)
```

# 2.2. Normalizing Rows and Columns
Often, we wish to normalize value of a row to within a specific range.

```python
x = np.array([ [ 0, 3, 4 ],
               [ 1, 6, 4 ] ])
```

```python
x_norm = np.amax(x, axis=1)
print(x_norm)
```

```python
x / x_norm
```

```python
print(x.shape, x_norm.shape)
```

```python
x_norm = x_norm.reshape((2, 1))
```

```python
x / x_norm
```

```python
x_norm
```

```python
# We can also use another normalization method 
x = np.array([ [ 0, 3, 4 ],
               [ 1, 6, 4 ] ])

# no need to reshape again 
x_norm = np.linalg.norm(x, ord = 2, axis = 1, keepdims = True)  # keepdims ensure the orignal shape
x / x_norm
```

```python
x_norm
```

# 2.4. Vectorization

```python
dim = 10000   # try increasing this after running experiment
```

```python
A = np.random.rand(dim, dim)
B = np.random.rand(dim, dim)
```

```python
A.shape
```

```python
# Naive approach
def add_arrays(A, B):
    C = np.zeros(A.shape)

    for i in range(A[0].size):
        for j in range(A[1].size):
            C[i, j] = A[i, j] + B[i, j]
    return C
```

```python
%time C = add_arrays(A, B)
```

```python
%time C = A + B
```

```python
import time 

# Non-vectorized time 
start = time.time() 

C = add_arrays(A, B)

end = time.time() 
non_vec_time = end - start 


## Vectorized time 
start = time.time() 

C = A + B 

end = time.time() 
vec_time = end - start 
```

```python
(vec_time / non_vec_time) * 100
```

```python
vec_time
```

```python
non_vec_time
```

```python
%%time
total = 0
for i in np.arange(100_000_000): 
    total += i 
print(total)
```

```python
%time sum(np.arange(100_000_000))
```

Moral of the story: don't write your own code when performance is an issue.
