# 1.1. Vector Representation

```python
class Vector:
    def __init__(self, x = 0.0, y = 0.0):
        self.x = x
        self.y = y

    def __str__(self):
        return f"[{self.x}, {self.y}]"
```

```python
a = Vector(2, 4)
```

```python
print(a)
```

```python
b = Vector(5, 2)
print(b)
```

```python
def add(self, b):
    c = Vector()
    c.x = self.x + b.x
    c.y = self.y + b.y
    return c

Vector.add = add
```

```python
c = a.add(b)
print(c)
```

```python
# scalar multiplication
def mul(self, s):
    return Vector(s * self.x, s * self.y)

Vector.mul = mul
```

```python
d = a.mul(2)
print(d)
```

```python
def sub(self, b):
    # We use the defination of vector subtraction (from maths prespective)
    # instead of defining something new

    return self.add( b.mul(-1) )

Vector.sub = sub
```

```python
d_min_b = d.sub(b)
print(d_min_b)
```

```python
def dot_prod(self, b):
    return (self.x * b.x) + (self.y * b.y)

Vector.dot_prod = dot_prod
```

```python
d_dot_b = d.dot_prod(b)
print(d_dot_b)
```

```python
def cross_prod(self, b):
    return self.x * b.y - self.y * b.x

Vector.cross_prod = cross_prod
```

```python
d_cross_b = d.cross_prod(b)
print(d_cross_b)
```

# 1.2. Maatrix Representation (Dense)

```python
class Matrix:

    

    def __init__(self, dims, fill):
        """
        Constructor for creating a Matrix
        Parameters:
            dims: tuple containing size of both dimenstions of Matrxi ie (i, j) mean i x j sized Matrix
            fill: default value for all the elements of the Matrix, can be zero
    
        """
        self.rows = dims[0]
        self.cols = dims[1]
        self.A = [
            [fill] * self.cols        # no of elements in each rows
            for i in range(self.rows) # this many rows
        ]
        
```

```python
[2] * 4
```

```python
m = Matrix((3, 4), 2.0)
```

```python
print(m)
```

```python
def __str__(self):
    rows = len(self.A) # Get the first dimenstion
    ret = ''

    for i in range(rows):
        cols = len(self.A[i])

        for j in range(cols):
            ret += str(self.A[i][j]) + "\t"
        ret += "\n"

    return ret

Matrix.__str__ = __str__
```

```python
print(m)
```

```python
%time n = Matrix((100, 100), 0.0)
```

This is taking up so much memory because it's storing all the 0s <b>explicity!</b> This is called a dense \
representation of matrices. Another way to do this is to not store 0s at all! Which is called a sparse.

```python
# recall that we can get values from our matrix using indices 
def get(self, i, j): 
    
    # Error checking 
    if i < 0 or i > self.rows: 
        raise ValueError("Row index out of range.")
    if j < 0 or j > self.cols: 
        raise ValueError("Column index out of range.")
    
    # Value return
    return self.A[i][j]

Matrix.get = get
```

# 1.3. Matrix Representation (Sparse)

```python
class Matrix: 
 
    def __init__(self, dims):    
        self.rows = dims[0]  
        self.cols = dims[1]   
        self.vals = {} 
        
        # Let's assume for a minute that fill is 0 
        
    # obviously need a new __str__ here .... 
```

```python
def set(self, i, j, val): 
    self.vals[   (i, j)    ] = val 


Matrix.set = set 
```

```python
# sparse implementation of get 
def get(self, i, j): 
    
    # Error checking 
    if i < 0 or i > self.rows: 
        raise ValueError("Row index out of range.")
    if j < 0 or j > self.cols: 
        raise ValueError("Column index out of range.")
    
    
    # value return 
    if (i, j) in self.vals: 
        return self.vals[  (i, j)  ]
    
    return 0.0
    
Matrix.get = get
```

```python
m = Matrix((5, 5))

print(m.vals)
```

```python
m.get(1, 1)
```

```python
m.get(10, 2)
```

```python
m.set(1, 2, 15.0)
```

```python
m.get(1, 2)
```
