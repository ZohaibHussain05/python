# 1.1. Vector Representation

```python
class Vector:
    def __init__(self, x = 0.0, y = 0.0):
        self.x = x
        self.y = y

    def __str__(self):
        return f"[{self.x}, {self.y}]"
```

```python
a = Vector(2, 4)
```

```python
print(a)
```

```python
b = Vector(5, 2)
print(b)
```

```python
def add(self, b):
    c = Vector()
    c.x = self.x + b.x
    c.y = self.y + b.y
    return c

Vector.add = add
```

```python
c = a.add(b)
print(c)
```

```python
# scalar multiplication
def mul(self, s):
    return Vector(s * self.x, s * self.y)

Vector.mul = mul
```

```python
d = a.mul(2)
print(d)
```

```python
def sub(self, b):
    # We use the defination of vector subtraction (from maths prespective)
    # instead of defining something new

    return self.add( b.mul(-1) )

Vector.sub = sub
```

```python
d_min_b = d.sub(b)
print(d_min_b)
```

```python
def dot_prod(self, b):
    return (self.x * b.x) + (self.y * b.y)

Vector.dot_prod = dot_prod
```

```python
d_dot_b = d.dot_prod(b)
print(d_dot_b)
```

```python
def cross_prod(self, b):
    return self.x * b.y - self.y * b.x

Vector.cross_prod = cross_prod
```

```python
d_cross_b = d.cross_prod(b)
print(d_cross_b)
```
